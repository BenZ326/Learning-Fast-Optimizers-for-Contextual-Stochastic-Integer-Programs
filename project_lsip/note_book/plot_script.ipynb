```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.load("env_gap.npy")
ev_policy = np.load("ev_policy.npy")
ev_random = np.load("ev_random.npy")
ev_sqdist = np.load("eval_sqdist.npy")
ev_scip = np.load("ev_scip.npy")
loss = np.load("loss_init.npy")
```

```python
rp = []
for i in range(len(ev_scip)):
    rp.append((ev_scip[i]-ev_policy[i])/ev_scip[i])
```

```python
plt.plot(rp)
plt.title("Relative Percentage Between SCIP and The Policy")
plt.xlabel("Instance")
plt.ylabel("Relative Percentage")
plt.savefig("relative_with_baseline.png")
```

```python
plt.plot(loss)
plt.title("Loss of the NADE policy")
plt.xlabel("# Epochs")
plt.ylabel("Loss")
plt.savefig("Loss_withbaseline.png")
```

```python
plt.plot(ev_sqdist)
plt.title("Square Distance Between SCIP and The Policy")
plt.xlabel("# Evaluations")
plt.ylabel("Square Distance")
plt.savefig("SquareDistance_withbaseline")
```
